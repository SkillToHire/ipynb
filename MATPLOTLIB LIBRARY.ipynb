# MATPLOTLIB

* Matplotlib is an open source python visualization library
* This is a powerful plotting library useful for those working with Python and Numpy
* The module used in Matplotlib is pyplot

```python
#import matplotlib library
import matplotlib.pyplot as plt
```

## BASICS

### Simple plot

```python
x=[1,2,7,9,10]
plt.plot(x)
plt.show()
```

### Plot with x and y values

```python
#simple plot

x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.show()
```

```python
#other way to plot

plt.plot([1,2,3],[1,4,9])
plt.show()
```

## ADD TITLE AND AXIS LABELS

```python
#using title,xlabel,ylabel

x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title("our first plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## ADD MULTIPLE PLOTS

```python
#to draw multiple lines on same plot

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y)
plt.plot(x,z)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.show()
```

## ADD LEGEND FOR PLOTS

```python
#using legend will make us able to differentiate between two lines by using different color

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y)
plt.plot(x,z)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["y","z"])
plt.show()
```

```python
#other way to mention labels of legend

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y')
plt.plot(x,z,label='z')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

## COLOR,FONTS AND PLOT REPRESENTATION

```python
#using color
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r')
plt.plot(x,z,label='z',color='k')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

```python
#marker
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') #circle marker
plt.plot(x,z,label='z',color='k',marker='s')#square marker
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

```python
#we can change the color of the marker using markerfacecolor

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b') #circle marker
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g')#square marker
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

## LINESTYLE AND WIDTH

```python
linestyle =>{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
linewidth =>float
```

```python
#using linestyle
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b',linestyle='--') 
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g',linestyle='-.')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

```python
#using linewidth

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b',linewidth=5)
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g',linewidth=10)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

## SETTING X AND Y AXES LIMITS

```python
#we set x and y axes limits using xlim and ylim

x=[30,10,20]
y=[40,50,60]
z=[30,40,50]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b') 
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g')
plt.xlim(10,50)
plt.ylim(10,50)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

## SETTING UP GRIDS IN OUR PLOTS

```python
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#using linewidths for grids

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=5)
plt.show()
```

```python
#we can add color to the grid

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c')
plt.show()
```

```python
#we can add linestyle to the grid

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c',linestyle='-.')
plt.show()
```

## SAVING PLOTS

```python
syntax:
savefig(fname, dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)
#fname=filename
```

```python
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c',linestyle='-.')
plt.savefig('abc',dpi=300) #dots per inch #higher the dpi,higher is the resolution of the plot
plt.show() #here 'abc.png' will be saved to our current folder
```

```python
#uisng facecolor in the savefig command

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c',linestyle='-.')
plt.savefig('abc',dpi=300,facecolor='blue') #when we view the png file the backgroundcolor of the figure will be blue 
plt.show() #here 'abc.png' will be saved to our current folder
```

## SUBPLOTS

```python
from numpy import *
plt.subplot(211)

x=linspace(-10,10,150)
cosine_wave=np.cos(x)
sine_wave=np.sin(x)

plt.plot(x,cosine_wave,color='r',linewidth=2.5)

plt.subplot(212)
plt.plot(x,sine_wave,color='g',linewidth=3.5)
```

```python
#change the size of figure using 'figsize'

plt.figure(figsize=(10,5))
plt.subplot(211)

x=linspace(-10,10,150)
cosine_wave=np.cos(x)
sine_wave=np.sin(x)

plt.plot(x,cosine_wave,color='r',linewidth=2.5)

plt.subplot(212)
plt.plot(x,sine_wave,color='g',linewidth=3.5)
```

```python
#to view plots in parallel by changing the rows and cols

plt.subplot(121)

x=linspace(-10,10,150)
cosine_wave=np.cos(x)
sine_wave=np.sin(x)

plt.plot(x,cosine_wave,color='r',linewidth=2.5)

plt.subplot(122)
plt.plot(x,sine_wave,color='g',linewidth=3.5)
```

```python
# we can use axes method instead of subplot where in we can customize the location of the plot

plt.axes([-0.4,1,1,1]) #left,bottom,width,height

x=linspace(-10,10,150)
cosine_wave=np.cos(x)
sine_wave=np.sin(x)

plt.plot(x,cosine_wave,color='r')

plt.axes([1,1,1,1])
plt.plot(x,sine_wave,color='g')
```

```python
plt.axes([-0.4,1,1,1]) #left,bottom,width,height

x=linspace(-10,10,150)
cosine_wave=np.cos(x)
sine_wave=np.sin(x)

plt.plot(x,cosine_wave,color='r')

plt.axes([1,1.4,1,1]) #change slightly and observe the plot
plt.plot(x,sine_wave,color='g')
```

### PIE CHARTS

```python
#simple pie chart

labels=["dogs","cats","horses","fishes"]
sizes=[60,45,20,10]

plt.pie(sizes,labels=labels)
plt.show()
```

```python
#example
firms=["Firm_A","Firm_B","Firm_C","Firm_D","Firm_E"]
market_share=[20,25,15,10,20]
Explode=[0,0.1,0,0,0] ##only "explode" the 2nd slice (i.e. 'Firm_B')
plt.pie(market_share,explode=Explode,labels=firms,shadow=True,startangle=40)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(title="List of Firms")
plt.show()
```

### STACK PLOTS

```python
## stack plots

a=[1,2,3,4,5]
b=[1,2,5,6,7]
c=[3,2,5,4,6]
d=[7,8,5,6,7]

e=np.vstack([b,c,d])
plt.stackplot(a,e,colors=["m","c","r"])

plt.title("stack plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## TEXT ANNOTATIONS

```python
population_in_village=[534,689,258,401,724,689,435,534,545,435,423,789,905,561]
ids=range(1,15)

y_coordinate=min(population_in_village)
x_coordinate=ids[population_in_village.index(y_coordinate)]

plt.plot(ids,population_in_village)
plt.xlabel("ids")
plt.ylabel("population in numbers")
plt.title("Text annotation example")
plt.annotate('lowest value',xy=(x_coordinate,y_coordinate),xytext=(5,300),arrowprops=dict(facecolor='black',shrink=0.05))
plt.show()                                                                                  #lower the shrink,bigger is the arrow size
```
